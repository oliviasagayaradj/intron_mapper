```python
intron_start_l = []
intron_end_l = []

seq_id_l = []


for ii in range(df.shape[0]):
    if df.iloc[ii]['type'] == 'gene':
        geneStart = df.iloc[ii]['start']
        geneEnd = df.iloc[ii]['end']
    if df.iloc[ii]['type'] == 'exon':
        exonStart = df.iloc[ii]['start']
        exonEnd = df.iloc[ii]['end']
        if exonStart != geneStart:
            intron_end_l.append(exonStart - 1)
            seq_id_l.append(df.iloc[ii]['seq_id'])
        if exonEnd != geneEnd:
            intron_start_l.append(exonEnd + 1)
```

```python
fasta_path = 'assemblies/10015.contigs.fasta'

intron_d = {seqid:[] for seqid in set(seq_id_l)}

for seq_id, intronStart, intronEnd in zip(seq_id_l, intron_start_l, intron_end_l):
    for record in SeqIO.parse(fasta_path, 'fasta'):
        if record.id.split('_')[1]==seq_id:
            intron_d[seq_id].append(record[intronStart:intronEnd])
```

```python
intron_coord_d = {seqid:[] for seqid in set(seq_id_l)}
for jj, end in enumerate(intron_end_l):
    intron_coord_d[seq_id_l[jj]].append((intron_start_l[jj], end))

for record in SeqIO.parse(fasta_path, 'fasta'):
    contig_id = record.id.split('_')[1]
    print(contig_id)
    intron_coord_l = intron_coord_d[contig_id]
    print(intron_coord_l)
    #get the intron sequences and put them in a list
    #repeat for all ecotype
    #save intron sequences into fasta file. one file for each ecotype
        
```
