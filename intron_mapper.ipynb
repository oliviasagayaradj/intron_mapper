```python
intron_start_l = []
intron_end_l = []

seq_id_l = []

strand_l = []


for ii in range(df.shape[0]):
    if df.iloc[ii]['type'] == 'gene':
        geneStart = df.iloc[ii]['start']
        geneEnd = df.iloc[ii]['end']
    if df.iloc[ii]['type'] == 'exon':
        exonStart = df.iloc[ii]['start']
        exonEnd = df.iloc[ii]['end']
        if exonStart != geneStart:
            intron_end_l.append(exonStart - 1)
            seq_id_l.append(df.iloc[ii]['seq_id'])
            strand_l.append(df.iloc[ii]['strand'])
        if exonEnd != geneEnd:
            intron_start_l.append(exonEnd + 1)
```

```python
#get the intron sequences and put them in a list
fasta_path = 'assemblies/10015.contigs.fasta'

intron_d = {seqid:[] for seqid in set(seq_id_l)}

for record in SeqIO.parse(fasta_path, 'fasta'):
    for seq_id, intronStart, intronEnd, strand in zip(seq_id_l, intron_start_l, intron_end_l, strand_l):
        contig_id = record.id.split('_')[1]
        if seq_id == contig_id:
            intron = record.seq[intronStart:intronEnd]
            if strand == '-':
                intron_d[contig_id].append(intron.reverse_complement())
            else:
                intron_d[contig_id].append(intron)
```
